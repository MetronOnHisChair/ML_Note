## 什么是智能

如果你读过罗素的《人工智能，一种现代方法》，你会发现人们对于智能如何定义的看法一直在变化，早先的人们将智能定义为一种基于符号的推理，每种符号可以用来代表不同的知识，通过逻辑运算，不同的符号能推导出其它符号，以此形成智能的过程，以此为基础，近代曾出现过不同的人工智能语言用于构建智能系统，比如70年代曾出现的prolog（其中包含的离散数学知识不是现在的重点）

那么人们要构建这样一个智能系统应该怎么办呢？假设你知道了怎样去推导，基于一种如下式所示的逻辑$$a\to b\\b\to c\\? a\to c,yes!$$问题就变成了，你如何将不同的知识以这种规则编入电脑里？以prolog为例：

编写事实：  

awesome(Taurus).

金牛座很棒

isTaurus(niuniu).

牛牛是金牛座


编写逻辑规则:

hooray(X):- isTaurus(X).

如果一个人是金牛座，高呼万岁🙋

那么我们运行hooray(niuniu),就会返回True。通过以上的方法，以前的人们将规则编入计算机中，构建了可以推导得出结论的专家系统。很容易发现，当一个系统的运作需要考虑的变量和规则非常多或者难以表达时，这种方法是非常累人的。

**以上Prolog的语法你暂不需要了解**，只是作为一个额外的信息分享，近代统计学习的兴起，让人们可以从大量的数据中通过统计学获取相关的特征和联系，而不用去编写复杂的模型，加上现代计算机计算能力的提升，包括GPU对并行计算的超快提速（主要体现在深度学习中）,这些我之后会分享，海量的数据让工业界能够在不同的领域开发出帮助研究以及提高生产力的应用，但基于统计的方法现在也开始显现出了自身对模型描述和学习能力的限制，归根结底，人们对事物的认识本就是一个螺旋上升的过程，现在新兴的图神经网络我个人认为是一个很好的方向，我个人也相信智能绝不仅仅是一种统计学现象。
