牛牛你好👋 

想象一件事，一个初到世界的婴儿，他对这个世界是怎么运作的并不了解，这时候，他去外面到处看，渐渐的他长大了，相同的事情看久了，他明白了如果发生A，可能会发生B，如果有些东西长得像，可能是一回事～他去做尝试，当遇见一种状况时，他做了一个反应，他并不知道做这个反应会导致什么结果，但是当他做完的时候，有可能结果并不好，遇见栏杆太高，他想跳过去，结果摔了一跤；家长让他在家，他不做作业看电视吃零食，于是被批评了；但也有可能结果很不错，他越过了栏杆，觉得自己有成为超人的潜质；他好好学习，期末考了高分，家长给他买了XBOX,于是他有经验了，在了解了这些模式之后，他渐渐的总结出了自己的结论，用来描述这些事情都是怎么运作的，称之为模型，而这个通过不断探索获得模型的过程，称之为这个BOY的学习。

这篇笔记所参考的教材暂定是清华出版社李航编写的《统计学习原理》以及周志华编写的《机器学习》，我会将自己认为有交流价值的部分选出分享，与其采取令人反感的填鸭式,一个接着一个堆1234知识点和算法的传授方法，我希望能以一种更加形象的方式先给予直观的映像，更多能引发对方法思想的感受，然后再分享数学的方法和实现，如果笔记中有错误，或者没有包含一些可以有的有意思的知识点，欢迎提出和指正。
